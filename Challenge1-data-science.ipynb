### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
Facturaciontotaltienda1=0
Facturaciontotaltienda2=0
Facturaciontotaltienda3=0
Facturaciontotaltienda4=0

precios1=list(tienda.Precio)
for precio in precios1:
  Facturaciontotaltienda1+=precio

precios2=list(tienda2.Precio)
for precio in precios2:
  Facturaciontotaltienda2+=precio

precios3=list(tienda3.Precio)
for precio in precios3:
  Facturaciontotaltienda3+=precio

precios4=list(tienda4.Precio)
for precio in precios4:
    Facturaciontotaltienda4+=precio

print(f"Ventas por tienda : tienda1: {Facturaciontotaltienda1}, tienda2: {Facturaciontotaltienda2}, tienda3:{Facturaciontotaltienda3}, tienda4: {Facturaciontotaltienda4}")
```

# 2. Ventas por categoría

```python
tienda['Categoría del Producto']


```

```python
categoriastienda1 = list(tienda['Categoría del Producto'])
categoriastienda2 = list(tienda2['Categoría del Producto'])
categoriastienda3 = list(tienda3['Categoría del Producto'])
categoriastienda4 = list(tienda4['Categoría del Producto'])

def Conteocategorias(cat):
  conteo = {}
  for categoria in cat:
    if categoria not in conteo:
      conteo[categoria]=0
    conteo[categoria] +=1
  return conteo


Categoriastiendauno = Conteocategorias(categoriastienda1)
Categoriastiendados = Conteocategorias(categoriastienda2)
Categoriastiendatres = Conteocategorias(categoriastienda3)
Categoriastiendacuatro = Conteocategorias(categoriastienda4)

print(f'tienda 1: {Categoriastiendauno}')
print(f'tienda 2: {Categoriastiendados}')
print(f'tienda 3: {Categoriastiendatres}')
print(f'tienda 4: {Categoriastiendacuatro}')
```

# 3. Calificación promedio de la tienda


```python
import math
calificaciontienda1 = list(tienda['Calificación'])
calificaciontienda2 = list(tienda2['Calificación'])
calificaciontienda3 = list(tienda3['Calificación'])
calificaciontienda4 = list(tienda4['Calificación'])

CalificacionPromediotienda1 = sum(calificaciontienda1)/len(calificaciontienda1)

CalificacionPromediotienda2 = sum(calificaciontienda2)/len(calificaciontienda2)

CalificacionPromediotienda3 = sum(calificaciontienda3)/len(calificaciontienda3)

CalificacionPromediotienda4 = sum(calificaciontienda4)/len(calificaciontienda4)

print(f'Calificación tienda 1: {CalificacionPromediotienda1}, Calificación tienda 2: {CalificacionPromediotienda2}, Calificación tienda 3: {CalificacionPromediotienda3}, Calificación tienda 4: {CalificacionPromediotienda4}')
```

# 4. Productos más y menos vendidos

```python
productostienda1 = tienda['Producto']
productostienda2 = tienda2['Producto']
productostienda3 = tienda3['Producto']
productostienda4 = tienda4['Producto']

def Conteoproductos(prod):
  conteo = {}
  for categoria in prod:
    if categoria not in conteo:
      conteo[categoria]=0
    conteo[categoria] +=1
  return conteo

cantidadproductostienda1 = Conteoproductos(productostienda1)

cantidadproductostienda2 = Conteoproductos(productostienda2)

cantidadproductostienda3 = Conteoproductos(productostienda3)

cantidadproductostienda4 = Conteoproductos(productostienda4)

Productomasvendidotienda1 = max(cantidadproductostienda1, key=cantidadproductostienda1.get)
Productomasvendidotienda2 = max(cantidadproductostienda2, key = cantidadproductostienda2.get)
Productomasvendidotienda3 = max(cantidadproductostienda3, key =cantidadproductostienda3.get)
Productomasvendidotienda4 = max(cantidadproductostienda4, key = cantidadproductostienda4.get)

Productomenosvendidotienda1 = min(cantidadproductostienda1, key =cantidadproductostienda1.get)
Productomenosvendidotienda2 = min(cantidadproductostienda2, key=cantidadproductostienda2.get)
Productomenosvendidotienda3 = min(cantidadproductostienda3, key=cantidadproductostienda3.get)
Productomenosvendidotienda4 = min(cantidadproductostienda4, key=cantidadproductostienda4.get)

print(f'Productos mas vendidos: Tienda 1: {Productomasvendidotienda1}, Tienda 2: {Productomasvendidotienda2}, Tienda 3: {Productomasvendidotienda3}, Tienda4: {Productomasvendidotienda4}')

print(f'Productos menos vendidos: Tienda 1: {Productomenosvendidotienda1}, Tienda 2: {Productomenosvendidotienda2}, Tienda 3: {Productomenosvendidotienda3}, Tienda4: {Productomenosvendidotienda4}')
```

#Informe Final — Recomendación de Venta para el Sr. Juan
##Introducción

El Sr. Juan administra actualmente cuatro tiendas y desea vender aquella con el menor desempeño, con el fin de obtener capital para invertir en un nuevo negocio.
Con este objetivo, se realizó un análisis integral de los datos históricos de ventas, complementado con visualizaciones que permiten comprender mejor el desempeño de cada tienda.

El análisis consideró los siguientes factores:

- Ingresos totales de cada tienda

- Categorías de productos más y menos vendidas

- Calificaciones promedio de los clientes

- Productos más y menos vendidos

- Coste de envío promedio por tienda

A partir de estos elementos, se elaboró la presente recomendación, sustentada en evidencia cuantitativa y en la interpretación de las tendencias observadas en los gráficos generados.

##Desarrollo
1. Ingresos Totales

Los ingresos son el indicador principal para evaluar el rendimiento global de cada tienda. Los resultados revelan lo siguiente:

- Tienda 1: 1,150,880,400

- Tienda 2: 1,116,343,500

- Tienda 3: 1,098,019,600

- Tienda 4: 1,038,375,700

Entre las cuatro, Tienda 4 registra la facturación acumulada más baja, lo que sugiere un menor aporte económico al negocio.

La visualización de distribución porcentual confirma este comportamiento: Tienda 4 representa la porción más pequeña del total consolidado.

2. Categorías de Productos Más y Menos Vendidas

Cada tienda presenta una distribución distinta de categorías. En general, todas mantienen presencia en categorías como Electrónicos, Muebles, Electrodomésticos y Juguetes, pero con variaciones en demanda.

Tienda 4 muestra una distribución similar a las demás, aunque sin dominar ninguna categoría clave de alta rotación, lo que limita su crecimiento comparativo.

3. Calificaciones Promedio de Clientes

Este indicador refleja la experiencia del cliente y su satisfacción general. Los promedios obtenidos son:

- Tienda 1: 3.98

- Tienda 2: 4.04

- Tienda 3: 4.05

- Tienda 4: 3.99

Aunque Tienda 4 no posee la calificación más baja, tampoco destaca positivamente.
Se encuentra en un punto intermedio, sin una ventaja clara que compense su menor desempeño en ingresos.

4. Productos Más y Menos Vendidos

El análisis de los productos con mejor y peor desempeño revela:

###Tienda 4

- Más vendido: Cama box

- Menos vendido: Guitarra eléctrica

Si bien su producto más vendido corresponde a una categoría de alto valor, la presencia de artículos con baja rotación (como la guitarra eléctrica) indica que la tienda podría estar enfrentando problemas de demanda o de alineación entre inventario y preferencias del cliente.

Este patrón se refleja en la gráfica de productos más vendidos por tienda, donde Tienda 4 no muestra una ventaja significativa respecto a las demás.

5. Coste de Envío Promedio

Los costes logísticos pueden influir en la rentabilidad general. Las cifras obtenidas son:

- Tienda 1: 26,018.61

- Tienda 2: 25,216.24

- Tienda 3: 24,805.68

- Tienda 4: 23,459.46

En este aspecto, Tienda 4 tiene el coste de envío promedio más bajo, lo cual constituye una fortaleza operativa.
Sin embargo, esta ventaja no se traduce en mejores ventas ni mayor rentabilidad.

##Conclusión y Recomendación

Con base en el análisis integral realizado y sustentado en las visualizaciones generadas, se concluye que la tienda que el Sr. Juan debería vender es:

- Tienda 4
##Justificación:

- Es la tienda con el ingreso total más bajo, posicionándola como la unidad con menor rendimiento económico.

- Su calificación promedio no destaca, quedando por debajo de Tienda 2 y Tienda 3, que presentan mejor percepción por parte de los clientes.

- A pesar de tener el menor coste de envío promedio, esta fortaleza no compensa la brecha significativa en ingresos frente a las otras tiendas.

En conjunto, Tienda 4 presenta las debilidades más relevantes y el menor aporte al negocio, por lo que resulta la mejor candidata para vender y destinar los recursos obtenidos a nuevas oportunidades de inversión.
